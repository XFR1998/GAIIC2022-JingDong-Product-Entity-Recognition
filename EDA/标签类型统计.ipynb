```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_excel('../datasets/preliminary_contest_datasets/train_data/train.xlsx')
df.info()
```

```python
df['bio_label'][0].split(' ')
```

```python
tlabel_dict = dict()
for labels in tqdm(df['bio_label']):
    labels = labels.split(' ')
    for lb in labels:
        if lb not in tlabel_dict.keys():
            tlabel_dict[lb] = 1
        else:
            tlabel_dict[lb] += 1
```

```python
label_dict = dict()
for i in tlabel_dict.keys():
    if i =='O':
        label_dict['O'] = tlabel_dict['O']
    else:
        k = i.split('-')[1]
        if k not in label_dict.keys():
            label_dict[k] = tlabel_dict[i]
        else:
            label_dict[k] = max(label_dict[k], tlabel_dict[i])
label_dict
```

```python
labels_sort = sorted(label_dict.items(), key=lambda x:x[1])
labels_sort
```

```python
sum_nums = 0
for i in labels_sort:
    sum_nums += i[1]

sum_nums
```
